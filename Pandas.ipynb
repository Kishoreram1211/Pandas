Inspecting a DataFrame

```python
#information about the DataFrame
df.info()
```

```python
#finding the number of missing values
df.isnull().sum()
```

```python
#counting the values based on labels
df.value_counts('Survived')
```

```python
#group the values based on mean
df.groupby('Survived').mean()
```

```python
#number of values
df.count()
```

```python
#mean of each column
df.mean()
```

```python
#description of the DataFrame
df.describe()
```

```python
#dropping a column
df.drop(columns='Name' , axis=1)
```

```python
#locating a row using the index value
df.iloc[4]
```

```python
#locating a particular column
df.iloc[:,1]
```
